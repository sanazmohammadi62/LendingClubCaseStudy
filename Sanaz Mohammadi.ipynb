```python
## dealing with numerical values missing data
n_clos = train_data[["pub_rec_bankruptcies","total_bal_ex_mort", "total_bc_limit"]]
for i in n_clos:
  train_data[i].fillna(train_data[i].mean(axis=0), inplace=True)
```

```python
##to drop columns from index idx to the end because there are a lot of missing data
columns_to_drop = train_data.columns[55:]
##Drop the columns
train_data = train_data.drop(columns_to_drop, axis=1)
```

```python
###visualization###
def bar_chart(col):
  Approved = train_data[train_data["loan_status"] == "Y"][col].value_counts()
  Disapproved = train_data[train_data["loan_status"] == "N"][col].value_counts()

  dfl = pd.DataFrame([Approved, Disapproved])
  dfl.index = ["Approved", "Disapproved"]
  dfl.plot(kind="bar")
```

```python
from sklearn.preprocessing import OrdinalEncoder

ord_enc = OrdinalEncoder
train_data[["loan_status"]] = ord_enc
train_data.head()
```
